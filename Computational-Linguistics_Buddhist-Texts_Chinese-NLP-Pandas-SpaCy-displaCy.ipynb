# Introduction

**Chinese NLP**
This proof-of-concept notebook uses a (Modern) Chinese SpaCy library as a base to tokenize and analyze parts of speech for a short Buddhist text. Mistakes in the tags were corrected by hand to train an improved model for Pre-Modern Literary Chinese Texts. It was built with Pandas, SpaCy, Xiaoquan Kong's Chinese SpaCy models. It implements a Part of Speech tagger and visualizations using diSplaCy. 

**Applying Standard Mandarin NLP Models & Historical and Regional Dialects and Languages**  
The Chinese SpaCy library uses Standard Mandarin to generate its tags for texts. However, the texts used by Chinese readers are not only written in modern Chinese. Chinese characters can be used to represent several different languages, including specialized language like Legal Chinese, Chinese “dialects” like Cantonese, Wu, Min, and Hakka, as well as pre-modern literary forms, like Classical Chinese, written vernacular Chinese, and poetic language. Chinese characters also historically served as the writing system for Korean, Japanese, and Vietnamese. All of these historical and current written languages differ from Standard Mandarin, some slightly and some greatly, in terms of language features like lexicon, morphology, syntax, and semantics. 

**Goals for Buddhist Chinese**  
Chinese Buddhist texts are generally written in a form of Classical Chinese, but they are still read by more than a quarter of a billion Buddhists throughout the modern Sinophone world[1]. This project demonstrates that the existing Chinese POS libraries are not yet sufficient to deal with these texts, as well as other texts written in Chinese characters that diverge from Standard Mandarin. This project is a first step towards correcting this gap.

[1] https://www.pewforum.org/2012/12/18/global-religious-landscape-buddhist/

## The Chinese Buddhist Canon

**Background**  
The Chinese Buddhist Canon is the source of texts for most branches of Chinese, Japanese, Korean, and Vietnamese Buddhism. It consists of a vast number of texts, making it a spectacular reservoir for corpus linguistics. One of the most widely used versions, the Taisho Tripitaka, contains more than 5,000 individual texts, with a character count totaling in the millions. The Canon contains not only Chinese translations of Indic scriptural texts, but also transliterations of incantations, and original Chinese works such as biographies of monks, annotated catalogues, commentaries on texts, and even texts that historical editors suspected were fake. Language-use in many of these different parts can vary.

**Scope of Analysis**  
This analysis focuses on the smallest analyzable set: a single sentence taken from one of the most popular - and shortest - texts, the Heart Sutra. We plan to expand to the whole sutra, then progress to a larger, popular sutra, the Lotus Sutra. The final goal is to train an AI with large enough samples from as many different sections as possible, so that it can recognize variations in linguistic patterns and eventually read the whole Canon.

## Related Studies

Because of its distinctive translation history over a long period of time, scholars have approached the Buddhist Canon as a source of research in a number of different ways. Scholars of Buddhist philosophy have used it to conduct philological inquiry into how Indic Buddhist concepts were translated into Chinese and understood by Chinese audiences. Historical linguists have used transliterations of Sanskrit in the Canon to inform reconstructions of Middle Chinese. More recently, particularly since the digitization of the canon, researchers have begun to approach it as a target for digital humanities and even machine learning. Some of the individuals who have initiated digital work on the Buddhist Canon include:

Marcus Bingenheimer Hung Jen-Jou, Zhang Bo-yong, and others affiliated with the Dharma Drum Institute of Liberal Arts, who have engaged in digitization of texts, data visualizations deriving from the texts, and applying computer-based methodologies to texts to answer humanistic and sociological questions.  

Charles Muller, who maintains one of the largest online dictionaries of Chinese Buddhist terms, including named entities and variants. 

Alex Amies, who has written web applications to analyze Chinese Buddhist texts.

# Part-of-Speech Tagging

**Applying the ZH Model**  
This project uses spaCy as the POS engine for the project. However, spaCy does not come with a Chinese character model by default. For that reason, this project builds on the work of Xiaoquan Kong who released a Chinese language spaCy model [on his Github](https://github.com/howl-anderson/Chinese_models_for_SpaCy), which has done a lot of the initial legwork for Chinese language NLP projects yet to come.

## Text Loader

In the following **Text Loader**, the user is prompted to enter a Title and Characters into the text fields. To use the default _Heart Sutra_ simply leave the fields blank and press RETURN. Other Chinese texts, including the remainder of the _Heart Sutra_, can be found in the [Taisho Canon](http://tripitaka.cbeta.org/T08n0251_001) digitized on [CBETA](http://tripitaka.cbeta.org/T). 

```python
inputTitle = input("Title of Text [Enter for Default]: ")

while True:
    if inputTitle == '':
        inputTitle = 'Heart Sutra First Sentence'
        inputCharacters = "觀自在菩薩，行深般若波羅蜜多時，照見五蘊皆空，度一切苦厄。"
        break
    else:
        inputCharacters = input("Enter Characters: ")
        break

print()
print("Complete.")
print()
print(inputTitle)
print()
print(inputCharacters)
```

## Load Chinese 'ZH'  spaCy Model

In the following cell, spaCy loads Kong's model using the package name **ZH** which will then be used to analyze the text. ZH is then applied to input Characters collected in the previous cell. **ZH** extracts linguistic features from the **User's Input Characters**, including, parts of speech tags("**tag_**"), such as nouns, verbs, adjectives, and dependency relationships among the tokens. 

One difference between spaCy's supported language models versus Kong's ZH model is the absence of lemmatization. Chinese language has non-inflective morphology. In other words, characters cannot be transformed with prefixes or suffixes in the same way that inflective morphologies, including English, append affixes to root words. 

```python
nlp = spacy.load('zh')
doc = nlp(inputCharacters)
doc.user_data["title"] = inputTitle
title = str(doc.user_data["title"])
headers = [
    'text', 'lemma_', 'pos_', 'tag_', 'dep_', 'shape_', 'is_alpha', 'is_stop',
    'has_vector', 'ent_iob_', 'ent_type_', 'vector_norm', 'is_oov'
]

print()
print("Title: " + title)
print()
print(doc)
print()
print("Linguistic Features:")
print(headers)
```

## Tokenize

After loading the document, spaCy's default tokenizer called to split the characters on whitespace. The resulting Doc object is list of tokens. 

```python
headers = [
    'text', 'lemma_', 'pos_', 'tag_', 'dep_', 'shape_', 'is_alpha', 'is_stop',
    'has_vector', 'ent_iob_', 'ent_type_', 'vector_norm', 'is_oov'
]

doc_data = []

for token in doc:
    token_data = [
        token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
        token.shape_, token.is_alpha, token.is_stop, token.has_vector,
        token.ent_iob_, token.ent_type_, token.vector_norm, token.is_oov
    ]
    doc_data.append(token_data)

display(pd.DataFrame(doc_data, columns=headers))
```

## Visualize Each Sentence

This snippet splits the loaded text by sentence at the "。" symbol. Each sentence is then visualized using **displaCy**, a tool for producing dependency trees according to relationships between parts of speech.

```python
sentences = inputCharacters.split('。')
count = 0
for sentence in sentences:
    count += 1
    doc = nlp(sentence)
    doc.user_data["title"] = str(count)

    headers = [
        'text', 'lemma_', 'pos_', 'tag_', 'dep_', 'shape_', 'is_alpha',
        'is_stop', 'has_vector', 'ent_iob_', 'ent_type_', 'vector_norm',
        'is_oov'
    ]

    doc_data = []

    for token in doc:
        token_data = [
            token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop, token.has_vector,
            token.ent_iob_, token.ent_type_, token.vector_norm, token.is_oov
        ]
    doc_data.append(token_data)

    pd.DataFrame(doc_data, columns=headers)
    displacy.render(doc, style="dep", jupyter=True, options={'distance': 125})
    svg = displacy.render(doc, style="dep")
    path = ("./zh-exports/_" + str(count) + "_" + title + ".svg")
    with open(path, "w+", encoding="utf-8") as f:
        f.write(svg)
        print("Exported " + path)
```

# Analysis

**Explaining ZH Outputs**  
As we can see from the charts above, the automatic tagging based on the SpaCy library is able to tokenize and assign POS tags to the sentence, as well as a create a dependency tree based on the tags to model the syntax. However, the results generated from the automatic tagging do not render the correct reading of the sentence. In fact, the automatic tagging contains a number of issues that generate an erroneous reading. First, the library does not recognize that the first five characters of the text are a single token, the named entity Avalokitesvara Bodhisattva. The computer also fails to recognize, and thus incorrectly tokenizes the characters for the word prajnaparamita, a frequently-mentioned Buddhist concept. The computer tokenizes imperfectly, but better, in the second half of the sentence, but the POS tags it generates for the tokens are suspect.
